```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('che.jpg', -1)

if __name__ == '__main__':
    img_gary = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(img_gary, 50, 150, apertureSize=3)
    orgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    op = orgb.copy()
    for line in cv2.HoughLinesP(edges, 1, np.pi / 180, 1, minLineLength=100, maxLineGap=10):
        x1, y1, x2, y2 = line[0]
        cv2.line(orgb, (x1, y1), (x2, y2), (0, 0, 255), 5)
    plt.subplot(121)
    plt.imshow(op)
    plt.axis('off')
    plt.subplot(122)
    plt.imshow(orgb)
    plt.axis('off')
```
