```python
from matplotlib import pyplot as plt
import numpy as np
from numpy import ndarray
import cv2


def show_cv2_image(img_: ndarray) -> None:
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img_, cv2.COLOR_BGR2RGB))


if __name__ == '__main__':
    img = cv2.imread('Lenna.png')
    image = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    sobelX = cv2.Sobel(image, cv2.CV_64F, 1, 0)
    sobelY = cv2.Sobel(image, cv2.CV_64F, 0, 1)
    sobelX = np.uint8(np.absolute(sobelX))
    sobelY = np.uint8(np.absolute(sobelY))
    sobelCombined = cv2.bitwise_or(sobelX, sobelY)
    show_cv2_image(sobelCombined)
    output = cv2.seamlessClone(sobelCombined, img, None, (256, 256), cv2.NORMAL_CLONE)
    show_cv2_image(output)
```
