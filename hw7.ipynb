```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread("go.png", 0)

if __name__ == '__main__':

    _ret, thresh1 = cv2.threshold(img, 190, 255, cv2.THRESH_BINARY)

    _ret, thresh2 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY_INV)
    white_img = cv2.morphologyEx(thresh1, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8))
    white_img = cv2.dilate(white_img, np.ones((10, 10), np.uint8))

    black_img = cv2.morphologyEx(thresh2, cv2.MORPH_OPEN, np.ones((5, 5), np.uint8))
    black_img = cv2.erode(black_img, np.ones((4, 4), np.uint8))

    total_img = cv2.bitwise_not(cv2.bitwise_and(cv2.bitwise_not(black_img), cv2.bitwise_not(white_img)))
    plt.axis('off')
    plt.imshow(cv2.cvtColor(total_img, cv2.COLOR_BGR2RGB))

    print(len(cv2.findContours(total_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)[0]))

```
