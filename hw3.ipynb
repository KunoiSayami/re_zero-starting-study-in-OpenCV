```python
from matplotlib import pyplot as plt
import numpy as np
from numpy import ndarray
import cv2


def show_cv2_image(img_: ndarray) -> None:
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img_, cv2.COLOR_BGR2RGB))


if __name__ == '__main__':
    img = cv2.imread('Lenna.png')

    mean = 0.0
    std = 1.0
    noisy_img = img + np.random.normal(mean, std, img.shape)
    noisy_img_clipped = np.clip(noisy_img, 0, 255).astype(np.uint8)

    show_cv2_image(noisy_img_clipped)

    kernel_size = (5, 5)
    sigma = 1.5

    row, col, ch = noisy_img_clipped.shape

    img = cv2.GaussianBlur(noisy_img_clipped, kernel_size, sigma)

    plt.figure()
    show_cv2_image(img)
```
