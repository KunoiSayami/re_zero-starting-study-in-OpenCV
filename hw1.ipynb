```python
from matplotlib import pyplot as plt
import numpy as np
import cv2

if __name__ == '__main__':
    img = cv2.imread('qrcode.png')
    img_gary = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img_gary, cv2.COLOR_GRAY2RGB))
    plt.figure()
    canny = cv2.Canny(img_gary, 30, 150)
    plt.axis('off')
    plt.imshow(cv2.cvtColor(canny, cv2.COLOR_BGR2RGB))
```
