```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from typing import Generator

img = cv2.imread('Lenna.png', -1)


def compare_brightness() -> Generator[None, None, None]:
    met = (cv2.COLOR_BGR2HSV, cv2.COLOR_BGR2YCrCb)
    for method in met:
        m = cv2.cvtColor(img, method)
        hist_result_ = cv2.calcHist([m], [2 if cv2.COLOR_BGR2HSV else 0], None, [256], [0, 256])
        plt.plot(hist_result_, color='b' if method == cv2.COLOR_BGR2HSV else 'r')
        plt.xlim([0, 256])
        yield


if __name__ == '__main__':
    color = ('b', 'g', 'r')
    for channel, col in enumerate(color):
        hist_result = cv2.calcHist([img], [channel], None, [256], [0, 256])
        plt.plot(hist_result, color=col)
        plt.xlim([0, 256])
    plt.show()

    for _ in compare_brightness():
        pass
    plt.show()
```
