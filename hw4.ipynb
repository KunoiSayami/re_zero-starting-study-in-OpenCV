```python
from matplotlib import pyplot as plt
from numpy import ndarray
import cv2


def show_cv2_image(img_: ndarray) -> None:
    plt.axis('off')
    plt.imshow(cv2.cvtColor(img_, cv2.COLOR_BGR2RGB))


if __name__ == '__main__':
    img = cv2.imread('LS002452.jpg')

    show_cv2_image(img)

    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    h, s, v = cv2.split(hsv)

    hm = cv2.inRange(h, 5, 12)
    sm = cv2.inRange(s, 20, 170)

    out = cv2.bitwise_and(img, img, mask=hm & sm)

    img[hm & sm > 0] = (0, 255, 0)

    plt.figure()
    show_cv2_image(img)
```
